### KPI 3: EDA & Quarter-Wise Channel growth-rate analysis 

```python
Q3 = pd.read_csv(r"C:\Users\12896\OneDrive\Desktop\Preparation\Maven\SQL\MA TOY FACTORY PROJECT\MA TOY FACTORT CSV\Q3_total_orders_from_search_types_and_utm_campaigns.csv")
```

```python
Q3['quarter_wise'] = Q3['quarter_wise'].map(map_val)
```

```python
Q3['date'] = pd.to_datetime(Q3['date'])
```

```python
Q3
```

```python

# Create the figure and primary axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot `Gsearch_nonbrand` on the primary y-axis
sns.lineplot(data=Q3, x='date', y='Gsearch_nonbrand', marker='o', label='Gsearch_nonbrand', color='blue', ax=ax1)
ax1.set_ylabel('Gsearch_nonbrand Volume', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')

# Create a secondary y-axis for `Bsearch_nonbrand`
ax2 = ax1.twinx()
sns.lineplot(data=Q3, x='date', y='Bsearch_nonbrand', marker='o', label='Bsearch_nonbrand', color='orange', ax=ax2)
ax2.set_ylabel('Bsearch_nonbrand Volume', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')
ax2.legend(loc='upper right')

# Add a title and x-axis label
plt.title('Comparison of Gsearch_nonbrand and Bsearch_nonbrand Over Quarters')
ax1.set_xlabel('Quarter')

# Format the x-axis for better readability
plt.xticks(rotation=45, ha='right')

# Final layout adjustments
sns.despine()  # Removes unnecessary chart borders
plt.tight_layout()
plt.show()

```

```python
data_2 = Q3.loc[:, ['date','organic_search', 'direct_type_in']]
```

```python

data_2 = Q3.loc[:, ['date', 'organic_search', 'direct_type_in']]

# Convert 'date' column to datetime for proper plotting
data_2['date'] = pd.to_datetime(data_2['date'])

# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Organic Search on the primary y-axis
sns.lineplot(
    data=data_2,
    x='date',
    y='organic_search',
    label='Organic Search',
    marker='o',
    ax=ax1,
    color='blue'
)
ax1.set_ylabel('Organic Search', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Plot Direct Type-in on the secondary y-axis
ax2 = ax1.twinx()
sns.lineplot(
    data=data_2,
    x='date',
    y='direct_type_in',
    label='Direct Type-in',
    marker='o',
    ax=ax2,
    color='red'
)
ax2.set_ylabel('Direct Type-in', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add titles and labels
plt.title('Organic Search VS Direct Type-in Over Time', fontsize=14)
ax1.set_xlabel('Date')

# Adjust legend for both axes
# lines, labels = ax1.get_legend_handles_labels()
# lines2, labels2 = ax2.get_legend_handles_labels()
# ax1.legend(lines + lines2, labels + labels2, loc='upper left')

# Beautify the plot
plt.xticks(rotation=45, ha='right')
sns.despine(left=False, right=False)
plt.tight_layout()

# Show the plot
plt.show()
```
